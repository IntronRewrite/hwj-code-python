```python
import matplotlib.pyplot as plt

# Define the functions
def x_squared(x):
    return x ** 2

def abs_x(x):
    return abs(x)

def smooth_l1(x):
    if abs(x) < 1:
        return 0.5 * x ** 2
    else:
        return abs(x) - 0.5

# Generate x values
x_values = torch.linspace(-2.0, 2.0, steps=100)

# Compute y values for each function
y_squared = x_squared(x_values)
y_abs = abs_x(x_values)
y_smooth_l1 = torch.tensor([smooth_l1(x) for x in x_values])

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x_values.numpy(), y_squared.numpy(), label='x^2')
plt.plot(x_values.numpy(), y_abs.numpy(), label='|x|')
plt.plot(x_values.numpy(), y_smooth_l1.numpy(), label='Smooth L1')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of x^2, |x|, and Smooth L1 functions')
plt.grid(True)


# Save the plot as a PNG file
plt.savefig('comparison_plot.png')

# Show the plot
plt.show()
```
